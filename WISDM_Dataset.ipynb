```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load the CSV data
file_path = 'C:/Users/pavan/OneDrive/Documents/MS Project/Datasets/WISDM/WISDM_ar_v1.1_raw.csv'
df = pd.read_csv(file_path)

# Print the first few rows to confirm data loading
print("First few rows of the data:")
print(df.head())

# Convert all values in 'x', 'y', 'z' columns to strings first
df['x'] = df['x'].astype(str).str.replace('[^0-9.-]', '', regex=True)
df['y'] = df['y'].astype(str).str.replace('[^0-9.-]', '', regex=True)
df['z'] = df['z'].astype(str).str.replace('[^0-9.-]', '', regex=True)

# Convert columns to appropriate data types
df['x'] = pd.to_numeric(df['x'], errors='coerce')
df['y'] = pd.to_numeric(df['y'], errors='coerce')
df['z'] = pd.to_numeric(df['z'], errors='coerce')

# Drop rows with any remaining non-numeric values
df.dropna(inplace=True)

# Ensure 'timestamp' is included as the third feature
df = df[['user', 'activity', 'timestamp', 'x', 'y', 'z']]

# Check the size of the data after cleaning
print(f"Data shape after cleaning: {df.shape}")

# Ensure there is sufficient data for segmentation
if df.empty:
    raise ValueError("The dataframe is empty. Please check the data loading and preprocessing steps.")

# Segment the data
def segment_data(data, window_size=200, step_size=50):
    segments = []
    labels = []
    print(f"Starting segmentation with window_size={window_size}, step_size={step_size}")
    for start in range(0, len(data) - window_size, step_size):
        end = start + window_size
        segment = data.iloc[start:end, 3:].values
        label = data.iloc[start:end]['activity'].mode()[0]
        segments.append(segment)
        labels.append(label)
        if start % 1000 == 0:  # Print progress every 1000 steps
            print(f"Processed {start} segments")
    return np.array(segments), np.array(labels)

segments, labels = segment_data(df)

# Check the shape of the segments
print(f'Segments shape: {segments.shape}')
print(f'Labels shape: {labels.shape}')

# Flatten the segments if the shape is correct
if segments.ndim == 3:
    num_segments, num_steps, num_features = segments.shape
    segments_flat = segments.reshape(num_segments, num_steps * num_features)
else:
    raise ValueError(f'Unexpected shape for segments: {segments.shape}')

# Normalize the data
scaler = StandardScaler()
segments_flat = scaler.fit_transform(segments_flat)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(segments_flat, labels, test_size=0.2, random_state=42)

# Train the RandomForest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Evaluate the model
y_pred = clf.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(classification_report(y_test, y_pred))
```
